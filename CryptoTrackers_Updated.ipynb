## Analysis of BTCUSD , based on Volume and Return Ratio by Mohd Taha Abbas 200010045

```python
#importing the .csv file
df = pd.read_csv('D:\Bitstamp_BTCUSD_1h (1).csv')
```

```python
volume=df['Volume USD']  #separate the volume column into a new series
print(len(df)) # True standard distribution
print('True Standard Deviation of Data is USD ', np.std(volume , ddof=1))

plt.hist(volume , density =True);
```

```python
print('True mean of Data is USD ' ,np.mean(volume))  #true mean
```

```python
#Sampling Distribution
```

```python
def s_distr(N=10000 , k=1000):

    mean=[]
    variance=[]
    for x in range(N):
        s_v=np.random.choice(volume , size=k)
        mean.append(np.mean(s_v))
        variance.append(np.var(s_v , ddof=1))
    print('Predicted Mean USD ',np.mean(mean))
    plt.hist(mean , density =True);
    
    
interact(s_distr , N=(10,100000) , k=(2,100000))
```

```python
nrv=norm(4499277.337603658 , 7734536.902256386)
x=np.linspace(8e5,8e6, 100000)
plt.plot(x ,nrv.pdf(x))
```

```python
def var_distr(N=10000 , k=1000):  #sampling distribution of variance
    true_var=np.var(volume , ddof=1)
    v2=[]
    mean=[]
    variance=[]
    for x in range(N):
        s_v=np.random.choice(volume , size=k)
        variance.append(np.var(s_v , ddof=1)*(k-1)/true_var)
        v2.append(np.var(s_v , ddof=1))
    print('Predicted Variance  USD' ,np.mean(v2)) 
    print('Predicted Standard Deviation USD'  , np.sqrt(np.mean(v2)))
    plt.subplot(1,2,1)
    plt.hist(variance , density =True);
    plt.subplot(1,2,2)
    ch=chi2(k-1)
    x=np.linspace(min(variance) ,max(variance) ,10000)
    plt.plot(x , ch.pdf(x))
    

    
interact(var_distr , N=(10,100000) , k=(2,1000))
```

```python
def var_distr2(N=10000 , k=1000):
    v2=[]
    true_v=np.var(rr ,ddof=1)
    variance=[]
    for x in range(N):
        s_v=np.random.choice(rr , size=k)
        variance.append(np.var(s_v , ddof=1))
        v2.append(np.var(s_v , ddof=1)*(k-1)/true_v)
    print('Predicted Variance ' , np.mean(variance))
    plt.hist(v2 , density =True ,bins=20);
    
interact(var_distr2 , N=(10,100000) , k=(2,30))
```

```python
#confidence intervals  of mean 
def fun(n,alpha):
    nrv=norm(0,1)
    sample=np.random.choice(volume , size=n)
    m=np.mean(sample)
    std_volume=np.std(volume , ddof=1)
    print((1-alpha)*100 , '% Confidence Interval for true mean of Volume USD is')
    print(m-nrv.ppf(1-alpha/2)*std_volume/np.sqrt(n) ,',' ,m  + nrv.ppf(1-alpha/2)*std_volume/np.sqrt(n))
    mean=[]
    variance=[]
    N=10000
    k=1000
    for x in range(N):
        s_v=np.random.choice(volume , size=k)
        mean.append(np.mean(s_v))
    plt.figure(figsize=(20,10))
    plt.hist(mean , density =True);
    plt.axvline(m-nrv.ppf(1-alpha/2)*std_volume/np.sqrt(n) , 0 ,1 ,color='r')
    plt.axvline(m+nrv.ppf(1-alpha/2)*std_volume/np.sqrt(n) , 0 ,1 ,color='r') 
    
interact( fun , n=(2,100000) , alpha =[0.01 ,0.05])
```

```python
#confidence intervals  of mean  return ratio
def fun(n,alpha):
    nrv=norm(0,1)
    sample=np.random.choice(rr , size=n)
    m=np.mean(sample)
    std_rr=np.std(rr , ddof=1)
    print((1-alpha)*100 , '% Confidence Interval for true mean of Return Ratio is')
    print(m-nrv.ppf(1-alpha/2)*std_rr/np.sqrt(n)  ,m  + nrv.ppf(1-alpha/2)*std_rr/np.sqrt(n))
    mean=[]
    N=10000
    k=1000
    for x in range(N):
        s_v=np.random.choice(rr , size=k)
        mean.append(np.mean(s_v))
    plt.figure(figsize=(20,10))
    plt.hist(mean , density =True);
    plt.axvline(m-nrv.ppf(1-alpha/2)*std_rr/np.sqrt(n) , 0 ,1 ,color='r')
    plt.axvline(m+nrv.ppf(1-alpha/2)*std_rr/np.sqrt(n) , 0 ,1 ,color='r') 
    
interact( fun, n=(2,100000) , alpha =[0.01 ,0.05])
```

# Hypothesis test 1
* $Null$ $Hypothesis :$ The average return ratio, on an hourly basis, of Bitcoin is equal to that of Ethereum.$(\mu_1=\mu_2)$
* $Alternate$ $Hypothesis :$ The average return ratio, on an hourly basis, of Bitcoin is greter than that of Ethereum. $(\mu_1>\mu_2)$


```python
var1=np.var(df1['Return Ratio'])
var2=np.var(df2['Return Ratio'])
```

In this case,we know the variances of the populations which are the return ratios for each type of coin.Hence,we can use the approach of known variances.Due to the central limit theorem $\bar{X}$ - $\bar{Y}$ is a normal distribution for a large sample size with centre at $\mu_1 - \mu_2$ and standard deviation $\sqrt{\sigma_1^{2}/n +\sigma_2^{2}/m}$.Hence,

$$\frac{\bar{X} - \bar{Y}-(\mu_1 - \mu_2)}{\sqrt{\sigma_1^{2}/n +\sigma_2^{2}/m}} \sim N(0,1)$$

For a better understanding of the distribution of $\bar{X}$ - $\bar{Y}$,we can plot it empirically.Since we can find the true means here,we can compare the empirical and theoritical distributions.

```python
mu1=np.mean(df1['Return Ratio'])
mu2=np.mean(df2['Return Ratio'])
```

```python
n=100
m=100
rv=norm(loc=mu1-mu2,scale=np.sqrt(var1/n+var2/m))
```

```python
s=[]
for i in range(1000):
    s.append(np.mean(df1['Return Ratio'].sample(n,replace=True))-np.mean(df2['Return Ratio'].sample(m,replace=True)))
plt.hist(s,bins=50,density=True);
x=np.linspace(0.2,1.6,50)
plt.plot(x,rv.pdf(x),'r--');
```

We notice that it resembles the required normal distribution and hence the normal distribution is a good approximation for practical purposes.

The level of significance, $\alpha = 0.01$

```python
norm().ppf(0.99)
```

```python
a=np.mean(df1['Return Ratio'].sample(n,replace=True))-np.mean(df2['Return Ratio'].sample(m,replace=True))
b=np.sqrt(var1/n+var2/m)
a/b
```

The critical region is beyond 2.326 and since the data point lies in this region,the null hypothesis is rejected.The average return ratio for Bitcoin is indeed greater than Ethereum for our cause.This incentivises investors to buy Bitcoin over Ethereum.

# Hypothesis Test 2

* $Null$ $Hypothesis :$ The average volume of trade(in USD), on an hourly basis, of Bitcoin is thrice that of Ethereum.$(\mu_{VT1}=3\mu_{VT2})$
* $Alternate$ $Hypothesis :$ The average volume of trade, on an hourly basis, of Bitcoin is greater than thrice that of Ethereum.$(\mu_{VT1}>3\mu_{VT2})$

In this case,we will use the p-value approach and assume unknown variances.The statistics used are $\bar{X}$ and $3\bar{Y}$.

We know that
$$\frac{\bar{X} - 3\bar{Y}-(\mu_1 - 3\mu_2)}{\sqrt{S_1^{2}/n +9S_2^{2}/m}} \sim t_{r}$$
$$r=\frac{(S_1^2/n+9S_2^2/m)^2}{\frac{(S_1/n)^2}{n-1}+\frac{(3S_2/m)^2}{m-1}}$$
Here,$S_1$ and $S_2$ are sample standard deviations of volume of trade of Bitcoin and Ethereum respectively.

```python
s1=df1['Volume USD'].sample(n,replace=True)
s2=df2['Volume USD'].sample(m,replace=True)
m1=np.mean(s1)
m2=np.mean(s2)
var1=np.var(s1)
var2=np.var(s2)
```

```python
a=(var1/n+9*var2/m)**2
b=(np.sqrt(var1)/n)**2/(n-1)+(3*np.sqrt(var2)/m)**2/(m-1)
r=a/b
```

```python
rvt=t(r)
valuet=(m1-3*m2)/(np.sqrt(var1/n+9*var2/m))
```

$\alpha = 0.05$

```python
pvalue=1-rvt.cdf(valuet)
pvalue
```

As p-value < $\alpha$ chosen here,we can reject the null hypothesis.The average trade volume of Bitcoin is more than thrice that of Ethereum, giving an indication of investors' preference for Bitcoin over Ethereum.

# Hypothesis Test 3

* $Null$ $Hypothesis :$ The average fluctuation, on an hourly basis, of Bitcoin is equal to that of Ethereum.$(\mu_1=\mu_2)$
* $Alternate$ $Hypothesis :$ The average fluctuation, on an hourly basis, of Bitcoin is lesser than that of Ethereum. $(\mu_1<\mu_2)$
 
 Here,the fluctuation is measured by $\frac{high-low}{low}$
 
 We use the known varince case for simplicity.

$$\frac{\bar{X} - \bar{Y}-(\mu_1 - \mu_2)}{\sqrt{\sigma_1^{2}/n +\sigma_2^{2}/m}} \sim N(0,1)$$

```python
ser1=pd.Series((df1['high']-df1['low'])/df1['low'])
ser2=pd.Series((df1['high']-df1['low'])/df1['low'])
```

```python
sig1=np.std(ser1)
sig2=np.std(ser2)
```

```python
a=np.mean(ser1.sample(n,replace=True))-np.mean(ser2.sample(m,replace=True))
b=np.sqrt(sig1**2/n+sig2**2/m)
a/b
```

$\alpha = 0.1$

```python
norm().ppf(0.1)
```

The critical region corresponds to the region lesser than -1.28155 and since we get a point in the critical region,we reject the null hypothesis with a type 1 error of 10%.This hypothesis gives some insight on the stability and predictibility of Bitcoin.

```python
res=np.linalg.lstsq(a,s2, rcond=None)
coeff=res[0]
coeff
```

```python
plt.scatter(s1,s2)
vp=coeff[0]*s1+coeff[1]
plt.plot(s1,vp,'r');
```

# Coefficient of Determination($R^2$)

- $SSE = \sum (\hat{y}_i - y_i)^2$
- $SST = \sum (y_i - \bar{y})^2$
- $SSR = SST - SSE$
- $R^2 = SSR/SST$

```python
sse = np.sum((vp - s2)**2)
vbar = np.mean(s2)
sst = np.sum((s2 - vbar)**2)
ssr = sst - sse
rsq = ssr/sst
rsq
```

```python
plt.plot(s1, s2, 'o')
plt.plot(s1, vp, 'r', linewidth=2, label='Least Squares Fit $(R^2 = %.3f)$' % rsq)
plt.legend(fontsize=12)
plt.grid()
```

We see that the coefficient of determination is not large enough to support our hypothesis.Hence,our simple linear model does not encompass all the factors that affect the price-volume of trade relationship.

We can try to fit in a higher degree curve to see if it fits in better.

```python
un=np.polyfit(s1,s2,3)
un
```

```python
plt.scatter(s1,s2)
vps=un[0]*s1**3+un[1]*s1**2+un[2]*s1+un[3]
plt.plot(s1,vps,'r');
```

```python
plt.scatter(s1,s2)
plt.plot(s1,vps,'r');
plt.ylim(0,40000000);
```
